```python
import pandas as pd
import prepare as ppr
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./spam_clean.csv')
```

```python
df['clean_text'] = [ppr.clean(text) for text in df.text]
```

```python
df.head()
```

```python
ham = df[df['label']=='ham']
spam = df[df['label']=='spam']
```

```python
from wordcloud import WordCloud
```

```python
all_text = ''
for text in spam.clean_text:
    all_text += text + ' '
```

```python
img = WordCloud(background_color='white').generate(all_text)
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
all_text = ''
for text in ham.clean_text:
    all_text += text+ ' '
img = WordCloud(background_color='white',random_state=123).generate(all_text) 
plt.imshow(img)
plt.axis('off')
```
